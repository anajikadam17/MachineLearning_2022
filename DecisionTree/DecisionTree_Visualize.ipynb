## Decision Tree Classification

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred, y_test))
```

## Decision Tree Visualization using dtreeviz library

```python
!pip install dtreeviz
```

```python
import graphviz.backend as be

from sklearn.datasets import load_iris
from dtreeviz.trees import *
from IPython.display import Image, display_svg, SVG
```

```python
iris =  load_iris()

X = iris.data
y = iris.target
print(X.shape)
print(y.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth=2)

clf.fit(X,y)
clf1 = DecisionTreeClassifier(max_depth=None)

clf1.fit(X,y)
```

```python
from sklearn.tree import plot_tree
plot_tree(clf)
```

Visualize by Dtreeviz

```python
viz = dtreeviz(clf,
               X,
               y,
               feature_names=iris.feature_names,
               class_names=["setosa","versicolor","verginica"],
               scale=2
               )
viz
```

```python
# Horizontal graph
viz = dtreeviz(clf1,
               X,
               y,
               feature_names=iris.feature_names,
               class_names=["setosa","versicolor","verginica"],
               scale = 2,
               orientation='LR'
               )
viz
```

```python
XX = iris.data[np.random.randint(0, len(iris.data)),:]  #Random Point
# Check a Point path
viz = dtreeviz(clf1,
               X,
               y,
               feature_names=iris.feature_names,
               class_names=["setosa","versicolor","verginica"],
               scale = 2,
               X = XX
               )
viz
```

```python
# Show node labels
XX = iris.data[np.random.randint(0, len(iris.data)),:]  #Random Point
# Check a Point path
viz = dtreeviz(clf1,
               X,
               y,
               feature_names=iris.feature_names,
               class_names=["setosa","versicolor","verginica"],
               scale = 2,
               show_node_labels=True,
               X = XX
               )
viz
```

```python
# fancy = false
viz = dtreeviz(clf1,
               X,
               y,
               feature_names=iris.feature_names,
               class_names=["setosa","versicolor","verginica"],
               scale = 1,
               fancy = False
               )
viz
```

## Decision tree Regressor for Regression Problems

```python
# from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor

import pandas as pd
import numpy as np


data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]

print(X.shape, y.shape)

reg = DecisionTreeRegressor(max_depth=2)
reg2 = DecisionTreeRegressor(max_depth=None)
```

```python
reg.fit(X,y)
reg2.fit(X,y)
```

```python
viz = dtreeviz(reg,
               X,
               y,
               target_name="price",
               scale = 2
               )
viz
```

```python
viz = dtreeviz(reg2,
               X,
               y,
               target_name="price",
               scale = 2
               )
viz
```
